```python
"""--------------------------------Data Integrity Check-----------------------------------"""


df = pd.read_sql("""SELECT 
    c.contract_id,
    c.customer_id,
    c.quote_id,
    q.status AS quote_status

FROM contracts c
JOIN cpq_quote q 
    ON c.quote_id = q.quote_id

WHERE q.status IN ('Draft', 'Approved', 'Cancelled')
""", con=engine)
df
```

So its Conclude that 0 quote_id whose status was "draft", "Approved", "Cancelled" are in the contract. 

Now we will check for the valid quote_id which should in the contract data.
Converted quote → MUST have contract  
If not → broken pipeline

```python
df = pd.read_sql("""SELECT 
    q.quote_id,
    q.customer_id,
    q.status

FROM cpq_quote q
LEFT JOIN contracts c 
    ON q.quote_id = c.quote_id

WHERE q.status = 'Converted'
  AND c.contract_id IS NULL;
""", con=engine)

df
```

```python
"""----------------------------------Referential integrity validation--------------------------------------------"""


df = pd.read_sql("""SELECT 
    c.contract_id,
    c.customer_id,
    c.quote_id

FROM contracts c
LEFT JOIN cpq_quote q 
    ON c.quote_id = q.quote_id

WHERE q.quote_id IS NULL; """, con=engine)
df
```

```python
"""----------------------------------Duplicate records check--------------------------------------------"""
"""-------------------------------------cpq_quote table-------------------------------------"""
df = pd.read_sql("""SELECT 
    quote_id,
    COUNT(*) AS cnt
FROM cpq_quote
GROUP BY quote_id
HAVING COUNT(*) > 1;""", con=engine)
df
```

Now we check weather the dicount which is given in the quote does is match the contract dicount table or not if not
those contract will be marked.


```python
"""-----------------------pricing consistency check between quote and contract-----------------------------"""

df = pd.read_sql("""SELECT 
    c.contract_id,
    c.customer_id,
    c.quote_id,

    q.discount_percent AS quote_discount,
    c.discount_percent AS contract_discount,

    CASE 
        WHEN q.discount_percent = c.discount_percent THEN 'MATCH'
        ELSE 'MISMATCH'
    END AS discount_flag

FROM contracts c
JOIN cpq_quote q 
    ON c.quote_id = q.quote_id

WHERE q.status = 'Converted'
  AND (q.discount_percent <> c.discount_percent 
       OR q.discount_percent IS NULL 
       OR c.discount_percent IS NULL);""", con=engine)
df
```

So this proof that in the contract discount is given the same as per the quote table show.

Customer 'C43' appears to have fully paid all invoices, with no outstanding balance observed in the collections data.

However, a deeper analysis reveals that revenue leakage is occurring at the billing stage, despite successful payment and collection.

At the quote level, a discount was applied with a defined start date but no corresponding end date.

This caused the discount to persist beyond its intended duration.

This incorrect discount configuration propagated through the system:

• Quote → discount applied  
• Contract → inherited discount without end date  
• Invoice → discounted amount generated every cycle  

As a result, invoices were generated at $820 instead of the expected $1000.

This led to a recurring revenue leakage of $180 per billing cycle.

Even though the customer paid the full invoiced amount, the company still incurred a loss due to incorrect billing logic.

This highlights that successful collections do not guarantee revenue correctness.

The issue is not in payment or collection, but in pricing and billing configuration.

Specifically, missing discount end dates can cause silent and recurring revenue leakage.
