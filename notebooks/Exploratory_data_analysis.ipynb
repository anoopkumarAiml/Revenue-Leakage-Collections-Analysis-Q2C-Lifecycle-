```python
""" Here first we bring the pricing_master and cpq_quote table together  
 to verify that discount which are given is within the range and if not those customer_id will be flagged"""

df = pd.read_sql("""SELECT 
    q.quote_id,
    q.customer_id,
    q.plan_id,

    q.discount_percent,
    p.max_discount_percent,

    CASE 
        WHEN q.discount_percent <= p.max_discount_percent 
        THEN 'VALID'
        ELSE 'VIOLATION'
    END AS status

FROM cpq_quote q
JOIN pricing_master p 
    ON q.plan_id = p.plan_id""", con=engine)
df

```

```python
result = pd.read_sql("""
SELECT 
    SUM(CASE WHEN q.discount_percent <= p.max_discount_percent THEN 1 ELSE 0 END) AS valid_count,
    SUM(CASE WHEN q.discount_percent > p.max_discount_percent THEN 1 ELSE 0 END) AS violation_count
FROM cpq_quote q
JOIN pricing_master p 
    ON q.plan_id = p.plan_id
""", con=engine)

result
df
df["status"].value_counts()
```

Let's see the quote vs contract. For this time we are going  to those customer who are converted to the contract. also we will verify that any coustomer with the wrong status enter to the contract and if yes then those customer will be Flagged.

```python
df = pd.read_sql("""SELECT * from cpq_quote""", con=engine)
df
df["status"].value_counts()

```

Draft      → nothing  
Approved   → waiting  
Converted  → NEW contract  
Amended    → MODIFY existing contract  
Rejected   → dead

Only those quote whose ststus is Converted and Amended will be in the contract and rest of them should not be present in the contract. We will verify if any quote_id whose status was draft , approved and rejected found in the contract then those quote_id will be flagged.



```python
"""--------------------------------Data Integrity Check-----------------------------------"""


df = pd.read_sql("""SELECT 
    c.contract_id,
    c.customer_id,
    c.quote_id,
    q.status AS quote_status

FROM contracts c
JOIN cpq_quote q 
    ON c.quote_id = q.quote_id

WHERE q.status IN ('Draft', 'Approved', 'Cancelled')
""", con=engine)
df
```

So its Conclude that 0 quote_id whose status was "draft", "Approved", "Cancelled" are in the contract. 

Now we will check for the valid quote_id which should in the contract data.
Converted quote → MUST have contract  
If not → broken pipeline

```python
df = pd.read_sql("""SELECT 
    q.quote_id,
    q.customer_id,
    q.status

FROM cpq_quote q
LEFT JOIN contracts c 
    ON q.quote_id = c.quote_id

WHERE q.status = 'Converted'
  AND c.contract_id IS NULL;
""", con=engine)

df
```

So all quote_id which are converted are present in the contract, now we check does all quote_id present in the contract does this exist in the quote table or not. If not then those quote_id present in the contract will be flagged.

```python
"""----------------------------------Referential integrity validation--------------------------------------------"""


df = pd.read_sql("""SELECT 
    c.contract_id,
    c.customer_id,
    c.quote_id

FROM contracts c
LEFT JOIN cpq_quote q 
    ON c.quote_id = q.quote_id

WHERE q.quote_id IS NULL; """, con=engine)
df
```

So its verify that there is no Orphan contract is created.
Now we check there is any duplicate exist or not in the quote and the contract table

```python
"""----------------------------------Duplicate records check--------------------------------------------"""
"""-------------------------------------cpq_quote table-------------------------------------"""
df = pd.read_sql("""SELECT 
    quote_id,
    COUNT(*) AS cnt
FROM cpq_quote
GROUP BY quote_id
HAVING COUNT(*) > 1;""", con=engine)
df
```

```python
"""----------------------------------Duplicate records check--------------------------------------------"""
"""-------------------------------------Contracts table-------------------------------------"""

df = pd.read_sql("""SELECT 
    contract_id,
    COUNT(*) AS cnt
FROM contracts
GROUP BY contract_id
HAVING COUNT(*) > 1;""", con=engine)
df
```

So there is No duplicate present in the quote and contract table.
Now we check weather the dicount which is given in the quote does is match the contract dicount table or not if not
those contract will be marked.


```python
"""-----------------------pricing consistency check between quote and contract-----------------------------"""

df = pd.read_sql("""SELECT 
    c.contract_id,
    c.customer_id,
    c.quote_id,

    q.discount_percent AS quote_discount,
    c.discount_percent AS contract_discount,

    CASE 
        WHEN q.discount_percent = c.discount_percent THEN 'MATCH'
        ELSE 'MISMATCH'
    END AS discount_flag

FROM contracts c
JOIN cpq_quote q 
    ON c.quote_id = q.quote_id

WHERE q.status = 'Converted'
  AND (q.discount_percent <> c.discount_percent 
       OR q.discount_percent IS NULL 
       OR c.discount_percent IS NULL);""", con=engine)
df
```

So this proof that in the contract discount is given the same as per the quote table show.
